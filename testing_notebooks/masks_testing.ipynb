```python
def random_transform(mask):
    mask = np.array(mask)
    mask = np.rot90(mask, np.random.randint(0, 4))
    if np.random.rand() > 0.5:
        mask = np.fliplr(mask)
    if np.random.rand() > 0.5:
        mask = np.flipud(mask)
    return mask
```

```python
# random transform the mask in order to change part of the mask to 0

mask_transformed = random_transform(mask)
```
